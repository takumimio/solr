```python
import json
from datetime import datetime
with open("/Users/hal/Desktop/accenture_bluetooth low energy_20140101-20140630.json") as json_file:
    keys = ["FirmName","TechName","Period"]
    json_data = json.load(json_file)
    json = json_data[0]
    #print(json)
    results = json['resultDict']
    length = len(results)
    #print(length)
    subdict = {}
    for key in keys:
        subdict[key]=json[key]
    
    
    period = subdict['Period']
    periods = period.split('-')
    start = periods[0]
    end = periods[1]
    start_object = datetime.strptime(start, '%Y/%m/%d')
    end_object = datetime.strptime(end, '%Y/%m/%d')
    subdict['start'] = start_object
    subdict['end'] = end_object
    subdict.pop('Period')
    
    
    for result in results:
        #print(subdict)
        #print(result)
        for key, value in result.items():
            #print(key, value)
            if key == 'date':
                date_object = datetime.strptime(value, '%Y-%m-%d')
                subdict[key] = date_object
            else:
                subdict[key] = value
        print(subdict)
        
        
        
   
```
